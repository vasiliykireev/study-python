# Задание 1

Загрузите из CSV-файла исходные данные о покупателях с веб-портала (до 15 строк). При необходимости преобразуйте данные. С помощью Seaborn и Plotly сформируйте диаграмму разброса зависимости возраста клиента от расходов (scatterplot).

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

clients = pd.read_csv("web_clients_correct.csv", nrows=16)

def scatterplot_seaborn(df, x, y):
    scatterplot = sns.scatterplot(data=clients, x="age", y="bill")
    return scatterplot
scatterplot_seaborn(clients, "age", "bill")

def scatterplot_plotly(df, x, y):
    scatterplot = px.scatter(df, x=x, y=y)
    return scatterplot.show()
scatterplot_plotly(clients, "age", "bill")
```
