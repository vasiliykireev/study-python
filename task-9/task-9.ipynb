# Задание 1
С помощью python смоделировать игру против лотерейного автомата типа «777».

Игрок платит 1 руб., после чего выпадает случайное целое число, равномерно распределенное от 0 до 999.
При некоторых значениях числа игрок получает выигрыш:
777: 200 рублей
999: 100 рублей
555: 50 рублей
333: 15 рублей
111: 10 рублей
**7: 3 рубля
*00: 2 рубля
**0: 1 рубль
(* – любая цифра)
Выгодна ли игра игроку?
Сколько в среднем приобретает или теряет игрок за одну игру?

```python
# Игра со ставочкой

import numpy
import re

class Game:
    reward = {
        '777': 200,
        '999': 100,
        '555': 50,
        '333': 15,
        '111': 10,
        r'..7': 3,
        r'.00': 2,
        r'..0': 1
    }

    def __init__(self):
        self._score = self.play()
        self._winnings = self.check_result(self._score)

    @property
    def score(self):
        """
        Выпавшие очки
        """
        return self._score

    @property
    def winnings(self):
        """
        Выигрыш
        """
        return self._winnings

    def play(self, debug=None):
        """
        Игра в автомат

        Аргументы:
        - debug (int): число для отладки

        Возвращает str: 3 цифры
        """
        if debug is not None:
            result = debug
            if debug > 999:
                return 'error'
        else:
            result = numpy.random.randint(0, 999)

        result = str(result)
        zeros_number = 3 - len(result)
        string = '0' * zeros_number
        result = string + result

        return result

    def check_result(self, score):
        """
        Проверка результата — сопоставление с вознаграждением из self.reward
        Если результат не сопоставился — возвращает 0

        Аргументы:
        - score(str): строка для сопоставления с вознаграждением

        Возвращает int: сумму выигрыша
        """
        for pattern, prize in self.reward.items():
            if re.fullmatch(pattern, score):
                return prize
        return 0
    
play = Game()
print(play.score)
print(play.winnings)
```
