# Задание 1
С помощью python смоделировать игру против лотерейного автомата типа «777».

Игрок платит 1 руб., после чего выпадает случайное целое число, равномерно распределенное от 0 до 999.
При некоторых значениях числа игрок получает выигрыш:
777: 200 рублей
999: 100 рублей
555: 50 рублей
333: 15 рублей
111: 10 рублей
**7: 3 рубля
*00: 2 рубля
**0: 1 рубль
(* – любая цифра)
Выгодна ли игра игроку?
Сколько в среднем приобретает или теряет игрок за одну игру?

```python
import re

# Стоимость одной игры
COST_PER_PLAY = 1
games = 1000

# Словарь вознаграждений
reward = {
    '777': 200,
    '999': 100,
    '555': 50,
    '333': 15,
    '111': 10,
    r'..7': 3,
    r'.00': 2,
    r'..0': 1
}

total_win = 0

# Перебираем все возможные числа от 0 до 999, потому что только так можно равновероятно получить все значения
for x in range(games):
    s = f"{x:03d}"  # строка с ведущими нулями, например 7 -> '007'
    win = 0
    for pattern, amount in reward.items():
        if re.fullmatch(pattern, s):
            win = amount
            break
    total_win += win

# Средний выигрыш и прибыль
avg_win = total_win / games
expected_gain = avg_win - COST_PER_PLAY

print(f"Общий выигрыш: {total_win}")
print(f"Общая прибыль: {total_win - COST_PER_PLAY*games}")
print(f"Средний выигрыш за игру: {avg_win:.2f} руб.")
print(f"Средняя прибыль (выигрыш - стоимость): {expected_gain:.2f} руб.")
print(f"Игра {'выгодна' if expected_gain > 0 else 'невыгодна'} для игрока.")
```

```python
# Игра со ставочкой

import numpy
import re
import time

class Game:
    reward = {
        '777': 200,
        '999': 100,
        '555': 50,
        '333': 15,
        '111': 10,
        r'..7': 3,
        r'.00': 2,
        r'..0': 1
    }

    def __init__(self, bet = 1):
        self.bet = bet
        self._winnings = self.check_result(self.play(), bet)

    @property
    def winnings(self):
        """
        Выигрыш
        """
        return self._winnings

    def play(self, debug=None):
        """
        Игра в автомат

        Аргументы:
        - debug (int): число для отладки

        Возвращает str: 3 цифры
        """
        digits = ['', '', '']
        for p in range(100):
            if p < 50:
                digits[0] = self.random_digit()
                digits[1] = self.random_digit()
                digits[2] = self.random_digit()
            if p >= 50 and p < 75 and p % 2 == 0:
                digits[1] = self.random_digit()
                digits[2] = self.random_digit()
            if p >= 75 and p % 4 == 0:
                digits[2] = self.random_digit()
            result = ''.join(digits)
            print(f"\r{result}", end="", flush=True)
            time.sleep(0.05)
        return result

    def random_digit(self):
        digit = str(numpy.random.randint(0,9))
        return digit

    def check_result(self, score, bet):
        """
        Проверка результата — сопоставление с вознаграждением из self.reward
        Если результат не сопоставился — возвращает 0

        Аргументы:
        - score(str): строка для сопоставления с вознаграждением

        Возвращает int: сумму выигрыша
        """
        for pattern, prize in self.reward.items():
            if re.fullmatch(pattern, score):
                winning = prize * bet
                return winning
        return 0

play = Game()
print("\nВыигрыш: ", play.winnings)
```
